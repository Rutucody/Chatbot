```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

df = pd.read_csv("https://media.geeksforgeeks.org/wp-content/uploads/Wine.csv")
X = df.drop('Customer_Segment', axis=1)
y = df['Customer_Segment']

plt.figure(figsize=(10,6))
for i in range(1, 4):
    plt.scatter(X[y == i]['Alcohol'], X[y == i]['Malic_Acid'], label=f'Customer Segment {i}', alpha=0.7)
plt.xlabel('Alcohol')
plt.ylabel('Malic_Acid')
plt.legend()
plt.title('Scatter Plot (Original Data)')
plt.show()

pca_df = pd.DataFrame(PCA(n_components=2).fit_transform(StandardScaler().fit_transform(X)), columns=['Principal Component 1', 'Principal Component 2'])
pca_df['Customer_Segment'] = y
print(pca_df)

plt.figure(figsize=(10, 6))
for i in range(1, 4):
    plt.scatter(pca_df[pca_df['Customer_Segment'] == i]['Principal Component 1'], pca_df[pca_df['Customer_Segment'] == i]['Principal Component 2'], label=f'Customer Segment {i}', alpha=0.7)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.title('PCA: Wine Dataset')
plt.show()

```
